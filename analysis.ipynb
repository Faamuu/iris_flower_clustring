**Overview:**

This data contains two variables "Hours"(independent) and "Scores"(dependent).

*Task:*

Predict the percentage of an student based on the no. of study hours.

What will be predicted score if a student studies for 9.25 hrs/day?


Data can be found at http://bit.ly/w-data

**As this task is based on only two variables so, by the thinks simple I'm also using simple methods**

```python
#import important libraries
import pandas as pd
import seaborn as sns
import numpy as np 
import matplotlib.pyplot as plt
```

```python
#load the data
df=pd.read_csv("Students_data.txt")
```

```python
#Check the relation between feature and label
sns.scatterplot(data=df,x='Hours',y="Scores")
```

```python
#Split data feature(X) and label(y) from given data
X=df['Hours']
y=df['Scores']
```

```python
# find co-efficienints of the equation
np.polyfit(X,y,1)
```

```python
hours=X
```

```python
#Here's the required equation
scores=(9.77580339*hours)+2.48367341
```

```python
sns.scatterplot(data=df,x='Hours',y='Scores')
plt.plot(X,scores,color='red')
```
